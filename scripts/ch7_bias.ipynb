```python
# Imports
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

from itertools import product
from functools import partial
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from ID3 import RandomizedID3Classifier, RandomizedID3Ensemble

def feature_importances(X, y, cls, n_trees=500):
    clf = cls(n_estimators=n_trees).fit(X, y)
    
    if isinstance(clf, RandomizedID3Ensemble):
        imp = np.sum(clf.feature_importances_, axis=1)
        
    else:
        imp = np.zeros(X.shape[1])
        
        for tree in clf.estimators_:
            imp += tree.tree_.compute_feature_importances(normalize=False)
            
        imp = imp / n_trees
        
    return imp
```

```python
def generate_or(n1=2, n2=20):
    X = np.array([v for v in product(range(n1), range(n2))]).astype(np.int32)
    y = (X[:, 0] >= n1/2) | (X[:, 1] >= n2/2)
    return X, y

def generate_copy(n1=2, n2=20):
    X = np.array([np.arange(n2), np.arange(n2)]).T
    X[:, 0] = X[:, 1] >= n2/2
    y = X[:, 0]
    return X, y
```

```python
r = {}
g = generate_or

for name, cls in [("ID3", partial(RandomizedID3Ensemble, base_estimator=RandomizedID3Classifier(k=1))),
                  ("ETs", partial(ExtraTreesClassifier, max_features=1, criterion="entropy")),
                  ("ETs max_features=2", partial(ExtraTreesClassifier, max_features=2, criterion="entropy")),
                  ("RF bootstrap=True", partial(RandomForestClassifier, max_features=1, bootstrap=True, criterion="entropy")),
                  ("RF bootstrap=True max_features=2", partial(RandomForestClassifier, max_features=2, bootstrap=True, criterion="entropy")), 
                  ("RF bootstrap=False", partial(RandomForestClassifier, bootstrap=False, criterion="entropy")),
                  ("RF bootstrap=False max_features=2", partial(RandomForestClassifier, max_features=2, bootstrap=False, criterion="entropy"))]:
    f = []
    for n2 in range(2, 10+1, 2):
        X, y = g(n1=2, n2=n2)
        f.append(feature_importances(X, y, cls=cls))
    r[name] = np.array(f)
```

```python
for name, f in sorted(r.items()):
    plt.plot(range(2, 10+1, 2), f[:, 0], "o-", label="X1")
    plt.plot(range(2, 10+1, 2), f[:, 1], "o-", label="X2")
    plt.ylim([0., 1.0])
    plt.title(name)
    plt.legend(loc="best")
    plt.show()
```

```python
def generate_strobl(n_samples=120):
    X = np.array([v for v in product(range(2), range(4), range(10), range(20), range(2))]).astype(np.int32)
    X, y = X[:, :-1], X[:, -1]
    indices = np.random.permutation(X.shape[0])[:n_samples]
    return X[indices], y[indices]
```

```python
X, y = generate_strobl()
r = {}

for name, cls in [("ID3", partial(RandomizedID3Ensemble, base_estimator=RandomizedID3Classifier(k=1))),
                  ("ETs", partial(ExtraTreesClassifier, max_features=1, criterion="entropy")),
                  ("ETs max_features=2", partial(ExtraTreesClassifier, max_features=2, criterion="entropy")),
                  #("RF bootstrap=True", partial(RandomForestClassifier, max_features=1, bootstrap=True, criterion="entropy")),
                  #("RF bootstrap=True max_features=2", partial(RandomForestClassifier, max_features=2, bootstrap=True, criterion="entropy")), 
                  ("RF bootstrap=False", partial(RandomForestClassifier, bootstrap=False, criterion="entropy")),
                  ("RF bootstrap=False max_features=2", partial(RandomForestClassifier, max_features=2, bootstrap=False, criterion="entropy"))]:
    f = []
    f.append(feature_importances(X, y, cls=cls))
    r[name] = np.array(f)
```
